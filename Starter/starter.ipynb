```python
import os 
from litellm import completion

os.environ['GEMINI_API_KEY'] = ""
```

```python
def generate_response(msg):
    """Call LLM to get response"""

    messages =[
    {"role": 'system', 'content': 'You are someone to help me to familiar the use of interact with LLM via api'},
    {'role': 'user', 'content': msg}
]

    response = completion(
        model = "gemini/gemini-2.5-flash",
        messages=messages,
        max_tokens=4096
    )
    return response
```

```python
generate_response("What are the available primitive data types in python?").choices[0].message.content
```
